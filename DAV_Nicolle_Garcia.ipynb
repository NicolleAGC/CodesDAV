# Preprocessing

```python
import pandas as pd

# Load your data
dataset_river = pd.read_csv('/content/drive/MyDrive/DAV/historical_river_lam_tsuen.csv')

# Display data types of each column
print(dataset_river.dtypes)
```

```python
import pandas as pd


# Columns to convert from 'object' to numerical
columns_to_convert = [
    '5-Day Biochemical Oxygen Demand (mg/L)',
    'Ammonia-Nitrogen (mg/L)',
    'Nitrate-Nitrogen (mg/L)',
    'Orthophosphate Phosphorus (mg/L)',
    'Silica (as SiO₂) (mg/L)',
    'Suspended solids (mg/L)',
    'Total Kjeldahl Nitrogen (mg/L)',
    'Total Phosphorus (mg/L)',
    'Total Solids (mg/L)',
    'Total Volatile Solids (mg/L)'
]

# Convert columns to numeric, setting errors='coerce' to replace non-convertible values with NaN
for column in columns_to_convert:
    dataset_river[column] = pd.to_numeric(dataset_river[column], errors='coerce')

# Check new data types to confirm conversion
print(dataset_river.dtypes)
print(dataset_river.head())



```

```python
#Renaming of columns

data = dataset_river.rename(columns={
    '5-Day Biochemical Oxygen Demand (mg/L)': 'BOD5',
    'Ammonia-Nitrogen (mg/L)': 'NH3-N',
    'Conductivity (μS/cm)': 'Conductivity',
    'Dissolved Oxygen (%saturation)': 'DO%',
    'Dissolved Oxygen (mg/L)': 'DO',
    'Nitrate-Nitrogen (mg/L)': 'NO3-N',
    'Orthophosphate Phosphorus (mg/L)': 'PO4-P',
    'pH': 'pH',
    'Silica (as SiO₂) (mg/L)': 'Silica',
    'Suspended solids (mg/L)': 'SS',
    'Total Kjeldahl Nitrogen (mg/L)': 'TKN',
    'Total Phosphorus (mg/L)': 'Total P',
    'Total Solids (mg/L)': 'TS',
    'Total Volatile Solids (mg/L)': 'TVS',
    'Turbidity (NTU)': 'Turbidity',
    'Water Temperature (°C)': 'Temp°C'
})

#check the new names
print(data.head())

#missing value verification
print(data.isna().sum())
print("\n")


```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

#Scaling data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.select_dtypes(include=[np.number]))

# Create of imputer object with 3 neighbors
imputer = KNNImputer(n_neighbors=3)

# Apply imputer on scaled data
data_imputed_scaled = imputer.fit_transform(data_scaled)

# Convert back the scaled imputed data to original scale
data_imputed = scaler.inverse_transform(data_imputed_scaled)

# Creating a DataFrame from the imputed data
data_imputed = pd.DataFrame(data_imputed, columns=data.select_dtypes(include=[np.number]).columns)

# If you have categorical columns handle them as needed, then concat with numerical columns
df = pd.concat([data_imputed, data.select_dtypes(exclude=[np.number])], axis=1)

# Check the imputation
print(df.isnull().sum())


df.to_csv('historical_river_lam_tsuen.csv', index=False)
```

Recursive Feature Elimination

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split


#Classify Model into category
def classify_nh3_n(value):
    if value < 0.5:
        return 1
    elif 0.5 <= value <= 1.0:
        return 2
    elif 1.0 < value <= 2.0:
        return 3
    elif 2.0 < value <= 5.0:
        return 4
    else:
        return 5

df['NH3-N_Rank'] = df['NH3-N'].apply(classify_nh3_n)

X = df.drop('NH3-N_Rank', axis=1)
y = df['NH3-N_Rank']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model
rf.fit(X_train, y_train)

# Feature importance
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")
for f in range(X_train.shape[1]):
    print("%d. feature %s (%f)" % (f + 1, X_train.columns[indices[f]], importances[indices[f]]))

# RFE
rfe = RFE(estimator=rf, n_features_to_select=6, step=1)
rfe.fit(X_train, y_train)

# Summarize the selection of the attributes
print('Selected features after RFE:', X_train.columns[rfe.support_])
```

```python
#New dataset for NH3-N analysis, dropping values with least correlation
features_to_drop = [
    'TS', 'DO%', 'DO', 'NO3-N', 'TVS', 'Silica',
    'SS', 'pH', 'Turbidity', 'Temp°C', 'Sample No'
]

# Drop these features from the dataframe
data_nh3n = df.drop(columns=features_to_drop)

# Print the columns of the new dataframe to verify
print(data_nh3n)

# Save the new dataset to a new CSV file
data_nh3n.to_csv('data_nh3n.csv', index=False)
```

```python
#KNN Classifier

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, roc_auc_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


features = df[['BOD5', 'NH3-N', 'TKN', 'PO4-P', 'Total P', 'Conductivity']]
target = df['NH3-N_Rank']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the kNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the classifier
knn_classifier.fit(X_train_scaled, y_train)

# Predict the test set results using probabilities for ROC curve
y_probs = knn_classifier.predict_proba(X_test_scaled)


# Create Evaluation metrics
y_pred = knn_classifier.predict(X_test_scaled)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')
accuracy = accuracy_score(y_test, y_pred)
roc_auc_ovr = roc_auc_score(y_test, y_probs, multi_class="ovr", average="macro")


# Print Confusion Matrix and Classification Report
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('\nClassification Report:\n', classification_report(y_test, y_pred))

# Creating a DataFrame to display metrics in a table
metrics_knn = pd.DataFrame({
    'Precision': [precision],
    'Recall': [recall],
    'F1 Score': [f1],
    'Accuracy': [accuracy],
    'ROC AUC (OVR)': [roc_auc_ovr]
})

print(metrics_knn)
```

```python
#SVR

import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/DAV/Lam_Tsuen_historical.csv')

# Convert string columns to category type and encode
data['Station'] = data['Station'].astype('category')
data['Station_codes'] = data['Station'].cat.codes
data.drop(columns=['Station'], inplace=True)

# Data preprocessing
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.select_dtypes(include=[np.number]))
imputer = KNNImputer(n_neighbors=3)
data_imputed_scaled = imputer.fit_transform(data_scaled)
data_imputed = scaler.inverse_transform(data_imputed_scaled)
data_imputed = pd.DataFrame(data_imputed, columns=data.select_dtypes(include=[np.number]).columns)
data= pd.concat([data_imputed, data.select_dtypes(exclude=[np.number])], axis=1)

#Dates was set for indexing
data['Dates'] = pd.to_datetime(data['Dates'])
data.set_index('Dates', inplace=True)

# Aggregating data
aggregated_data = data.groupby('Dates').mean()

#Forward filling in case any missing value remains
aggregated_data.ffill(inplace=True)

# Create lagged features for the time series data
variable = 'NH3-N'
for i in range(1, 4):  
    aggregated_data[f'lag_{i}'] = aggregated_data[variable].shift(i)
aggregated_data.dropna(inplace=True)  

# Define feature matrix (X) and target vector (y)
X = aggregated_data.drop(columns=[variable])
y = aggregated_data[variable]

# Split the dataset into train and test sets
split_index = int(len(aggregated_data) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the SVR model
svr = SVR(kernel='rbf')
svr.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = svr.predict(X_test_scaled)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mse = mean_squared_error(y_test, y_pred)

# Print evaluation metrics
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MSE: {mse}')

# Plotting the actual vs. the predicted values
plt.figure(figsize=(10, 5))
plt.plot(y_test.index, y_test, label='Actual', alpha=0.6)
plt.plot(y_test.index, y_pred, label='Predicted', color='red', alpha=0.6)
plt.title('SVR Forecasting - Actual vs. Predicted')
plt.legend()
plt.show()
```

```python
#XG Boost
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/DAV/Lam_Tsuen_historical.csv')

# Convert string columns to category type and encode
data['Station'] = data['Station'].astype('category')
data['Station_codes'] = data['Station'].cat.codes
data.drop(columns=['Station'], inplace=True)

# Data preprocessing
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.select_dtypes(include=[np.number]))
imputer = KNNImputer(n_neighbors=3)
data_imputed_scaled = imputer.fit_transform(data_scaled)
data_imputed = scaler.inverse_transform(data_imputed_scaled)
data_imputed = pd.DataFrame(data_imputed, columns=data.select_dtypes(include=[np.number]).columns)
data= pd.concat([data_imputed, data.select_dtypes(exclude=[np.number])], axis=1)

#Dates was set for indexing
data['Dates'] = pd.to_datetime(data['Dates'])
data.set_index('Dates', inplace=True)

# Aggregating data
aggregated_data = data.groupby('Dates').mean()

# forward-fill to handle if there is still any missing value
aggregated_data.ffill(inplace=True)

# Create lagged features for the time series data
variable = 'NH3-N'
for i in range(1, 4):  # Creating 3 lagged features for example
    aggregated_data[f'lag_{i}'] = aggregated_data[variable].shift(i)
aggregated_data.dropna(inplace=True)  # Drop rows with NaN values created by shifting


X = aggregated_data.drop(columns=[variable])  # Features excluding the target
y = aggregated_data[variable]  # Target variable

# Split the dataset into train and test sets
split_index = int(len(aggregated_data) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# XGBoost requires DMatrix
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Define XGBoost model parameters
xgb_params = {
    'eta': 0.1,
    'max_depth': 3,
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse'
}

# Train the model
bst = xgb.train(xgb_params, dtrain, num_boost_round=100)

# Predict on the test set
y_pred = bst.predict(dtest)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mse = mean_squared_error(y_test, y_pred)

# Print evaluation metrics
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MSE: {mse}')

# Plotting the actual vs. the predicted values
plt.figure(figsize=(10, 5))
plt.plot(y_test.index, y_test, label='Actual', alpha=0.6)
plt.plot(y_test.index, y_pred, label='Predicted', color='red', alpha=0.6)
plt.title('XGBoost Forecasting - Actual vs. Predicted')
plt.legend()
plt.show()
```

```python

#LSTM
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/DAV/Lam_Tsuen_historical.csv')

# Convert string columns to category type and encode
data['Station'] = data['Station'].astype('category')
data['Station_codes'] = data['Station'].cat.codes
data.drop(columns=['Station'], inplace=True)

# Data preprocessing
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.select_dtypes(include=[np.number]))

# Dates set for indexing
data['Dates'] = pd.to_datetime(data['Dates'])
data.set_index('Dates', inplace=True)

# Aggregating data
aggregated_data = data.groupby('Dates').mean()

#Forward filling in case any missing value remains
aggregated_data.ffill(inplace=True)

# Create lagged features for the time series data
variable = 'NH3-N'
n_lags = 4  # Number of lags and future steps (for simplicity, using same number for both)
for i in range(1, n_lags + 1):
    aggregated_data[f'lag_{i}'] = aggregated_data[variable].shift(i)
aggregated_data.dropna(inplace=True)

# Prepare input data for LSTM [samples, time steps, features]
X = aggregated_data[[f'lag_{i}' for i in range(1, n_lags + 1)]].values
y = aggregated_data[variable].values
X = scaler.fit_transform(X)  # Normalize features
X = X.reshape(X.shape[0], X.shape[1], 1)  # Reshaping for LSTM

# Split the dataset into train and test sets
split_index = int(len(X) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Define LSTM model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(n_lags, 1)),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')

# Fit model and capture history
history = model.fit(X_train, y_train, epochs=50, verbose=1, batch_size=32, validation_data=(X_test, y_test))

# Predict and evaluate
y_pred = model.predict(X_test).flatten()  # Flatten for easier handling

# Calculate evaluation metrics
mae = np.mean(np.abs(y_test - y_pred))
rmse = np.sqrt(np.mean((y_test - y_pred)**2))
mse = np.mean((y_test - y_pred)**2)

# Print evaluation metrics
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MSE: {mse}')

# Plotting the actual vs. the predicted values
plt.figure(figsize=(10, 5))
plt.plot(y_test, label='Actual')
plt.plot(y_pred, label='Predicted', color='red')
plt.title('LSTM Forecasting - Actual vs. Predicted')
plt.legend()
plt.show()

# Plot training & validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.title('Training and Validation Loss Per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#DNN
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix, log_loss
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/DAV/Paper/data_nh3n.csv')

# NH3-N_Rank set as the target
X = data.drop('NH3-N_Rank', axis=1).values
y = data['NH3-N_Rank'].values

# One-hot encode the labels
y_encoded = to_categorical(y - 1)  

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Define the deep neural network
model = Sequential([
    Dense(512, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(5, activation='softmax')  # Output layer with softmax for multi-class classification
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {accuracy*100:.2f}%")

# Convert probabilities to class predictions
y_pred_probs = model.predict(X_test)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Calculate metrics
f1 = f1_score(y_true_classes, y_pred_classes, average='macro')  
roc_auc = roc_auc_score(to_categorical(y_true_classes), y_pred_probs, multi_class='ovr') 
logloss = log_loss(y_test, y_pred_probs)
cm = confusion_matrix(y_true_classes, y_pred_classes)


# Print metrics
print(f"F1 Score: {f1}")
print(f"ROC AUC Score: {roc_auc}")
print(f"Log Loss: {logloss}")
print("Confusion Matrix:")
print(cm)

# Plot training & validation accuracy values
plt.figure(figsize=(6, 10))
plt.subplot(2, 1, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(2, 1, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

plt.tight_layout()
plt.show()
```
